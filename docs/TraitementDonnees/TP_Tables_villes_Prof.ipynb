## Fusion de deux bases de données

Les deux bases de données précéadentes peuvent être liées (fusionneés).

```python
villes_court=info_villes[['nom_ville','population_villes','altitude','geonameid_villes','iso']]
pays_et_capital=pandas.merge(pays,villes_court,left_on=('capital','iso'),right_on=('nom_ville','iso'))
pays_et_capital.to_csv('pays_et_capital.csv', sep = ';')
```
